```python
def precision_at_k(actual, predicted, k=5):
    predicted = predicted[:k]
    return len(set(predicted) & set(actual)) / k
```

```python
import numpy as np

def dcg(relevance):
    return sum([rel / np.log2(idx + 2) for idx, rel in enumerate(relevance)])

def ndcg_at_k(actual, predicted, k=5):
    predicted = predicted[:k]

    relevance = [1 if item in actual else 0 for item in predicted]
    ideal_relevance = sorted(relevance, reverse=True)

    dcg_val = dcg(relevance)
    idcg_val = dcg(ideal_relevance)

    return dcg_val / idcg_val if idcg_val > 0 else 0
```

## Phase 2


```python
import pandas as pd
import numpy as np
import pickle
import scipy.sparse
from scipy.sparse import csr_matrix
import implicit

# Load saved Phase 1 data
tmdb_with_ml_id = pd.DataFrame(pickle.load(open('tmdb_ml_dict.pkl', 'rb')))
ratings_final   = pd.read_csv('data/ratings_final.csv')

print("✅ Data loaded")
print(f"Movies  : {len(tmdb_with_ml_id):,}")
print(f"Ratings : {len(ratings_final):,}")
print(f"implicit version: {implicit.__version__}")
```

```python
# Only keep ratings for movies that exist in our TMDB dataset
# This is the key fix that aligns everything correctly

valid_movie_ids = set(tmdb_with_ml_id['movieId'].values.tolist())
ratings_clean   = ratings_final[
    ratings_final['movieId'].isin(valid_movie_ids)
].copy().reset_index(drop=True)

print(f"Ratings before filter : {len(ratings_final):,}")
print(f"Ratings after filter  : {len(ratings_clean):,}")
print(f"Removed               : {len(ratings_final) - len(ratings_clean):,}")
```

```python
# Build mappings from filtered ratings ONLY
# This guarantees indices always match

clean_user_ids  = ratings_clean['userId'].unique()
clean_movie_ids = ratings_clean['movieId'].unique()

user_to_idx  = {int(uid): idx for idx, uid in enumerate(clean_user_ids)}
movie_to_idx = {int(mid): idx for idx, mid in enumerate(clean_movie_ids)}
idx_to_movie = {idx: int(mid) for mid, idx in movie_to_idx.items()}
idx_to_user  = {idx: int(uid) for uid, idx in user_to_idx.items()}

ratings_clean['user_idx']  = ratings_clean['userId'].map(user_to_idx)
ratings_clean['movie_idx'] = ratings_clean['movieId'].map(movie_to_idx)

print("✅ Mappings built")
print(f"Unique users  : {len(clean_user_ids):,}")
print(f"Unique movies : {len(clean_movie_ids):,}")
print(f"Sample user IDs: {sorted(list(user_to_idx.keys()))[:5]}")
```

```python
# User x Movie (for recommendations)
sparse_user_movie = csr_matrix(
    (
        ratings_clean['rating'].astype(np.float32).values,
        (ratings_clean['user_idx'].values,
         ratings_clean['movie_idx'].values)
    ),
    shape=(len(clean_user_ids), len(clean_movie_ids))
)

# Movie x User (for ALS training)
sparse_movie_user = sparse_user_movie.T.tocsr()

print("✅ Sparse matrices built")
print(f"User  x Movie : {sparse_user_movie.shape}")
print(f"Movie x User  : {sparse_movie_user.shape}")
```

```python
# implicit ALS treats COLUMNS as items
# So we train on User x Movie directly
# and it will learn Movie factors from columns

sparse_user_movie = csr_matrix(
    (
        ratings_clean['rating'].astype(np.float32).values,
        (ratings_clean['user_idx'].values,
         ratings_clean['movie_idx'].values)
    ),
    shape=(len(clean_user_ids), len(clean_movie_ids))
)

sparse_movie_user = sparse_user_movie.T.tocsr()

print(f"User  x Movie : {sparse_user_movie.shape}")
print(f"Movie x User  : {sparse_movie_user.shape}")

# Train on User x Movie — implicit learns movie factors from columns
als_model = implicit.als.AlternatingLeastSquares(
    factors=100,
    iterations=20,
    regularization=0.1,
    random_state=42
)

print("\nTraining ALS... (3-8 minutes) 👇\n")
als_model.fit(sparse_user_movie)

print(f"\n✅ ALS trained!")
print(f"Item factors shape : {als_model.item_factors.shape}")
print(f"Movies count       : {len(clean_movie_ids)}")

if als_model.item_factors.shape[0] == len(clean_movie_ids):
    print("✅ Alignment verified!")
else:
    print(f"❌ Mismatch: {als_model.item_factors.shape[0]} vs {len(clean_movie_ids)}")
```

```python
def collab_recommend(user_id, n=10):
    user_id = int(user_id)

    # Unknown user — show popular movies
    if user_id not in user_to_idx:
        popular = (
            ratings_clean
            .groupby('movieId')['rating']
            .count()
            .sort_values(ascending=False)
            .head(n)
            .index.tolist()
        )
        results = []
        for mid in popular:
            row = tmdb_with_ml_id[tmdb_with_ml_id['movieId'] == mid]
            if len(row) > 0:
                results.append({
                    'title'   : row['title'].values[0],
                    'movie_id': int(row['movie_id'].values[0]),
                    'score'   : 'Popular'
                })
        return pd.DataFrame(results)

    user_idx = user_to_idx[user_id]

    # Pass movie_user matrix row for this user
    rec_ids, scores = als_model.recommend(
        userid=user_idx,
        user_items=sparse_user_movie[user_idx],
        N=n,
        filter_already_liked_items=True
    )

    results = []
    for movie_idx, score in zip(rec_ids, scores):
        movie_idx = int(movie_idx)
        if movie_idx not in idx_to_movie:
            continue
        ml_movie_id = idx_to_movie[movie_idx]
        row = tmdb_with_ml_id[tmdb_with_ml_id['movieId'] == ml_movie_id]
        if len(row) > 0:
            results.append({
                'title'   : row['title'].values[0],
                'movie_id': int(row['movie_id'].values[0]),
                'score'   : round(float(score), 4)
            })

    return pd.DataFrame(results)

print("✅ collab_recommend ready")
```

```python
print("Evaluating model... (1-2 minutes)\n")

test_users = sorted(list(user_to_idx.keys()))[:300]
hits  = 0
total = 0

for uid in test_users:
    user_idx = user_to_idx[uid]
    try:
        rec_ids, _ = als_model.recommend(
            userid=user_idx,
            user_items=sparse_user_movie[user_idx],
            N=10,
            filter_already_liked_items=False
        )

        actual_good = (
            ratings_clean[
                (ratings_clean['userId'] == uid) &
                (ratings_clean['rating'] >= 4.0)
            ]['movieId']
            .map(movie_to_idx)
            .dropna()
            .astype(int)
            .tolist()
        )

        hits  += len(set(rec_ids.tolist()) & set(actual_good))
        total += 1

    except:
        continue

precision = hits / (total * 10) if total > 0 else 0

print("=" * 45)
print(f"📊 ALS Evaluation Results")
print(f"   Precision@10 : {precision:.4f}")
print(f"   In English   : {precision*100:.1f}% of recommended")
print(f"   movies were ones users actually liked")
print(f"   Tested on    : {total} users")
print("=" * 45)
```

```python
import pandas as pd
import numpy as np
import pickle
import scipy.sparse
from scipy.sparse import csr_matrix
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
from nltk.stem.porter import PorterStemmer
import implicit

# Load TMDB content data
movies_dict     = pickle.load(open('movie_dict.pkl',     'rb'))
tmdb_with_ml_id = pd.DataFrame(pickle.load(open('tmdb_ml_dict.pkl', 'rb')))
movies          = pd.DataFrame(movies_dict)

# Load ALS collaborative model
als_model    = pickle.load(open('als_model.pkl',    'rb'))
user_to_idx  = pickle.load(open('user_to_idx.pkl',  'rb'))
movie_to_idx = pickle.load(open('movie_to_idx.pkl', 'rb'))
idx_to_movie = pickle.load(open('idx_to_movie.pkl', 'rb'))

# Load sparse matrix
sparse_user_movie = scipy.sparse.load_npz('sparse_user_movie.npz')

# Load clean ratings
ratings_clean = pd.read_csv('data/ratings_clean.csv')

print("✅ All files loaded")
print(f"Content movies  : {len(movies):,}")
print(f"Hybrid movies   : {len(tmdb_with_ml_id):,}")
print(f"Sparse matrix   : {sparse_user_movie.shape}")
```

```python
# Rebuild tags and similarity from your original preprocessing
ps = PorterStemmer()

def stem(text):
    return " ".join([ps.stem(word) for word in text.split()])

movies['tags'] = movies['tags'].apply(stem)

cv      = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(movies['tags']).toarray()

content_similarity = cosine_similarity(vectors)

print("✅ Content similarity matrix rebuilt")
print(f"Shape: {content_similarity.shape}")
```

```python
def get_content_scores(movie_title):
    """
    Returns a dictionary of {movie_title: similarity_score}
    for all movies based on content similarity.
    """
    if movie_title not in movies['title'].values:
        return {}

    movie_index = movies[movies['title'] == movie_title].index[0]
    distances   = content_similarity[movie_index]

    scores = {}
    for idx, score in enumerate(distances):
        title = movies.iloc[idx]['title']
        scores[title] = float(score)

    return scores

# Test it
test_scores = get_content_scores('Avatar')
top5 = sorted(test_scores.items(), key=lambda x: x[1], reverse=True)[1:6]

print("✅ Content score function ready")
print("\nTop 5 content matches for Avatar:")
for title, score in top5:
    print(f"   {title:<45} score: {score:.4f}")
```

```python
def get_collab_scores(user_id, n=50):
    """
    Returns a dictionary of {movie_title: collab_score}
    for top N movies based on ALS collaborative filtering.
    """
    user_id = int(user_id)

    if user_id not in user_to_idx:
        return {}

    user_idx = user_to_idx[user_id]

    rec_ids, scores = als_model.recommend(
        userid=user_idx,
        user_items=sparse_user_movie[user_idx],
        N=n,
        filter_already_liked_items=True
    )

    result = {}
    for movie_idx, score in zip(rec_ids, scores):
        movie_idx = int(movie_idx)
        if movie_idx not in idx_to_movie:
            continue
        ml_movie_id = idx_to_movie[movie_idx]
        row = tmdb_with_ml_id[tmdb_with_ml_id['movieId'] == ml_movie_id]
        if len(row) > 0:
            title = row['title'].values[0]
            result[title] = float(score)

    return result

# Test it
real_user = sorted(list(user_to_idx.keys()))[0]
test_collab = get_collab_scores(user_id=real_user, n=10)

print("✅ Collaborative score function ready")
print(f"\nTop collab picks for User {real_user}:")
for title, score in list(test_collab.items())[:5]:
    print(f"   {title:<45} score: {score:.4f}")
```

```python
def hybrid_recommend(movie_title, user_id=None, alpha=0.7, n=5):
    """
    Hybrid recommender combining content + collaborative filtering.
    
    Parameters:
    -----------
    movie_title : str   — movie the user selected
    user_id     : int   — logged in user (None = content only)
    alpha       : float — weight for content (0.7 = 70% content, 30% collab)
    n           : int   — number of recommendations to return

    Returns:
    --------
    DataFrame with title, content_score, collab_score, hybrid_score
    """

    # Step 1: Get content scores for all movies
    content_scores = get_content_scores(movie_title)

    if not content_scores:
        return pd.DataFrame({'error': [f'Movie "{movie_title}" not found']})

    # Step 2: Get collaborative scores if user is known
    collab_scores = {}
    if user_id is not None:
        collab_scores = get_collab_scores(user_id=user_id, n=100)

    # Step 3: Normalize content scores to 0-1 range
    max_content = max(content_scores.values()) if content_scores else 1
    min_content = min(content_scores.values()) if content_scores else 0
    range_content = max_content - min_content if max_content != min_content else 1

    # Step 4: Normalize collab scores to 0-1 range
    if collab_scores:
        max_collab   = max(collab_scores.values())
        min_collab   = min(collab_scores.values())
        range_collab = max_collab - min_collab if max_collab != min_collab else 1
    
    # Step 5: Combine scores
    all_titles = set(content_scores.keys())
    results    = []

    for title in all_titles:
        # Skip the selected movie itself
        if title == movie_title:
            continue

        # Normalized content score
        c_score = (content_scores.get(title, 0) - min_content) / range_content

        # Normalized collab score
        if collab_scores:
            raw_collab = collab_scores.get(title, 0)
            col_score  = (raw_collab - min_collab) / range_collab
        else:
            col_score = 0

        # Hybrid score
        if collab_scores:
            hybrid = alpha * c_score + (1 - alpha) * col_score
        else:
            hybrid = c_score  # content only if no user

        results.append({
            'title'         : title,
            'content_score' : round(c_score,  4),
            'collab_score'  : round(col_score, 4),
            'hybrid_score'  : round(hybrid,    4)
        })

    # Step 6: Sort by hybrid score and return top N
    results_df = (pd.DataFrame(results)
                  .sort_values('hybrid_score', ascending=False)
                  .head(n)
                  .reset_index(drop=True))

    return results_df

print("✅ hybrid_recommend function ready")
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
import pickle
import pandas as pd

# Load the BERT model
# 'all-MiniLM-L6-v2' is the perfect balance of speed and accuracy
# Used in production at many companies
print("Loading BERT model... (downloads ~90MB first time)")

bert_model = SentenceTransformer('all-MiniLM-L6-v2')

print("✅ BERT model loaded!")
print(f"Model: all-MiniLM-L6-v2")
print(f"Embedding size: 384 dimensions")
```

```python
# Load movies dataframe
movies_dict = pickle.load(open('movie_dict.pkl', 'rb'))
movies_bert = pd.DataFrame(movies_dict)

# BERT works better with natural readable text
# Unlike TF-IDF, BERT understands meaning so we don't need stemming
# We'll use the raw tags but make them more readable

def prepare_bert_text(tags):
    """
    Convert tags string into clean readable text for BERT.
    BERT understands natural language better than stemmed words.
    """
    # Replace common concatenated words back to readable form
    text = tags.replace('ScienceFiction', 'Science Fiction')
    text = text.replace('cultureclash', 'culture clash')
    text = text.replace('spacewar', 'space war')
    text = text.replace('basedonnovel', 'based on novel')
    return text.lower().strip()

movies_bert['bert_text'] = movies_bert['tags'].apply(prepare_bert_text)

print("✅ Text prepared for BERT")
print(f"Total movies: {len(movies_bert)}")
print(f"\nSample text for Avatar:")
print(movies_bert[movies_bert['title'] == 'Avatar']['bert_text'].values[0][:200])
```

```python
# This is the core step — BERT reads each movie's text
# and creates a 384-dimensional vector capturing semantic meaning

print("Generating BERT embeddings...")
print("This will take 3-8 minutes for 4800 movies\n")

texts = movies_bert['bert_text'].tolist()

# batch_size=64 processes 64 movies at once for speed
# show_progress_bar shows live progress
bert_embeddings = bert_model.encode(
    texts,
    batch_size=64,
    show_progress_bar=True,
    convert_to_numpy=True
)

print(f"\n✅ BERT embeddings generated!")
print(f"Shape: {bert_embeddings.shape}")
print(f"Meaning: {bert_embeddings.shape[0]} movies × {bert_embeddings.shape[1]} features")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

print("Computing BERT similarity matrix...")

# Compute cosine similarity between all movie embeddings
bert_similarity = cosine_similarity(bert_embeddings)

print(f"✅ BERT similarity matrix built")
print(f"Shape: {bert_similarity.shape}")

# Quick sanity check
avatar_idx  = movies_bert[movies_bert['title'] == 'Avatar'].index[0]
dark_knight = movies_bert[movies_bert['title'] == 'The Dark Knight'].index[0]
inception   = movies_bert[movies_bert['title'] == 'Inception'].index[0]

print(f"\nSanity check — similarity scores for Avatar:")
print(f"  Avatar vs Dark Knight : {bert_similarity[avatar_idx][dark_knight]:.4f}")
print(f"  Avatar vs Inception   : {bert_similarity[avatar_idx][inception]:.4f}")
print(f"  Avatar vs Avatar      : {bert_similarity[avatar_idx][avatar_idx]:.4f} (should be 1.0)")
```

```python
def bert_recommend(movie_title, n=5):
    """
    Returns top N recommendations using BERT embeddings.
    Understands semantic meaning — not just keyword matching.
    """
    if movie_title not in movies_bert['title'].values:
        return pd.DataFrame({'error': [f'Movie not found: {movie_title}']})

    movie_idx = movies_bert[movies_bert['title'] == movie_title].index[0]
    distances = bert_similarity[movie_idx]

    # Sort by similarity score
    similar_indices = sorted(
        list(enumerate(distances)),
        key=lambda x: x[1],
        reverse=True
    )[1:n+1]  # skip index 0 (itself)

    results = []
    for idx, score in similar_indices:
        results.append({
            'title'      : movies_bert.iloc[idx]['title'],
            'bert_score' : round(float(score), 4)
        })

    return pd.DataFrame(results)

print("✅ bert_recommend function ready")
```

```python
# This comparison is very important for your resume
# It shows BERT understands MEANING, not just keywords

test_movies = ['The Dark Knight', 'Inception', 'Avatar', 'The Godfather']

for movie in test_movies:
    if movie not in movies_bert['title'].values:
        continue

    # TF-IDF recommendations (your original system)
    movie_idx    = movies[movies['title'] == movie].index[0]
    tfidf_scores = content_similarity[movie_idx]
    tfidf_top5   = sorted(
        list(enumerate(tfidf_scores)),
        key=lambda x: x[1],
        reverse=True
    )[1:6]
    tfidf_titles = [movies.iloc[i[0]]['title'] for i in tfidf_top5]

    # BERT recommendations
    bert_top5   = bert_recommend(movie, n=5)['title'].tolist()

    print(f"\n{'='*55}")
    print(f"Movie: {movie}")
    print(f"{'='*55}")
    print(f"{'TF-IDF':<30} {'BERT':<30}")
    print(f"{'-'*55}")
    for t, b in zip(tfidf_titles, bert_top5):
        print(f"{t:<30} {b:<30}")
```

```python
def hybrid_recommend_bert(movie_title, user_id=None,
                          alpha=0.5, beta=0.3, gamma=0.2, n=5):
    """
    Full 3-layer hybrid recommender:

    Score = alpha × BERT_score
           + beta  × TF-IDF_score  
           + gamma × Collab_score

    Parameters:
    -----------
    alpha : weight for BERT        (default 0.5)
    beta  : weight for TF-IDF      (default 0.3)
    gamma : weight for ALS collab  (default 0.2)

    Note: alpha + beta + gamma should = 1.0
    """
    # Validate weights
    assert abs(alpha + beta + gamma - 1.0) < 0.01, \
        "Weights must sum to 1.0"

    # --- BERT scores ---
    if movie_title not in movies_bert['title'].values:
        return pd.DataFrame({'error': ['Movie not found']})

    bert_idx    = movies_bert[movies_bert['title'] == movie_title].index[0]
    bert_scores_raw = bert_similarity[bert_idx]

    # --- TF-IDF scores ---
    if movie_title in movies['title'].values:
        tfidf_idx        = movies[movies['title'] == movie_title].index[0]
        tfidf_scores_raw = content_similarity[tfidf_idx]
    else:
        tfidf_scores_raw = np.zeros(len(movies_bert))

    # --- Collab scores ---
    collab_dict = {}
    if user_id is not None:
        collab_dict = get_collab_scores(user_id=user_id, n=200)

    # --- Normalize helpers ---
    def normalize(arr):
        mn, mx = arr.min(), arr.max()
        if mx == mn:
            return arr
        return (arr - mn) / (mx - mn)

    bert_norm  = normalize(bert_scores_raw)
    tfidf_norm = normalize(tfidf_scores_raw) \
                 if len(tfidf_scores_raw) == len(movies_bert) \
                 else np.zeros(len(movies_bert))

    if collab_dict:
        max_c = max(collab_dict.values())
        min_c = min(collab_dict.values())
        rng_c = max_c - min_c if max_c != min_c else 1

    # --- Combine scores ---
    results = []
    for idx in range(len(movies_bert)):
        title = movies_bert.iloc[idx]['title']
        if title == movie_title:
            continue

        b_score = float(bert_norm[idx])
        t_score = float(tfidf_norm[idx]) \
                  if idx < len(tfidf_norm) else 0.0

        if collab_dict and title in collab_dict:
            c_score = (collab_dict[title] - min_c) / rng_c
        else:
            c_score = 0.0

        if user_id is not None:
            final = alpha * b_score + beta * t_score + gamma * c_score
        else:
            # No user — split between BERT and TF-IDF only
            final = (alpha / (alpha + beta)) * b_score + \
                    (beta  / (alpha + beta)) * t_score

        results.append({
            'title'       : title,
            'bert_score'  : round(b_score, 4),
            'tfidf_score' : round(t_score, 4),
            'collab_score': round(c_score, 4),
            'final_score' : round(final,   4)
        })

    return (pd.DataFrame(results)
            .sort_values('final_score', ascending=False)
            .head(n)
            .reset_index(drop=True))

print("✅ hybrid_recommend_bert function ready")
print("   Layers: BERT (50%) + TF-IDF (30%) + ALS (20%)")
```

```python
real_user = sorted(list(user_to_idx.keys()))[0]

print("=" * 60)
print("FULL 3-LAYER HYBRID TEST")
print("=" * 60)

test_movies = ['The Dark Knight', 'Inception', 'Avatar']

for movie in test_movies:
    print(f"\n🎬 Input movie : {movie}")
    print(f"👤 User        : {real_user}")
    print(f"{'-'*55}")

    recs = hybrid_recommend_bert(
        movie_title=movie,
        user_id=real_user,
        alpha=0.5,
        beta=0.3,
        gamma=0.2,
        n=5
    )

    print(f"{'Title':<35} {'BERT':>6} {'TFIDF':>6} "
          f"{'Collab':>7} {'Final':>6}")
    print("-" * 65)
    for _, row in recs.iterrows():
        print(f"{row['title']:<35} "
              f"{row['bert_score']:>6.3f} "
              f"{row['tfidf_score']:>6.3f} "
              f"{row['collab_score']:>7.3f} "
              f"{row['final_score']:>6.3f}")
```
